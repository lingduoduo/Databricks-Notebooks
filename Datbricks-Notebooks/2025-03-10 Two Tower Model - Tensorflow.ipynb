```python
# Updated Cell 2: Catalog Data & Expanded Triplet Generation

catalog_data = [
    {
        "title": "ADP Enhanced Insights",
        "actions": [
            "Manage Consent for ADP Benchmarks"
        ]
    },
    {
        "title": "Approval Delegation",
        "actions": [
            "All Delegates",
            "My Delegates Information"
        ]
    },
    {
        "title": "Approval Management dashboard",
        "actions": [
            "Approval Management Dashboard"
        ]
    },
    {
        "title": "Associate Data Import",
        "actions": [
            "Associate data import",
            "Core HR Inbound Import Staging List"
        ]
    },
    {
        "title": "Associate Document Management",
        "actions": [
            "Bulk Document Migration Dashboard",
            "Document Cloud Dashboard",
            "Document Management Dashboard"
        ]
    },
    {
        "title": "Associate External Identity Management",
        "actions": [
            "Assign Additional Associate Identifiers",
            "Configure Additional Associate Identifiers"
        ]
    },
    {
        "title": "Enhanced Workflows",
        "actions": [
            "Configure Enhanced Workflows",
            "Manage Enhanced Workflows"
        ]
    },
    {
        "title": "FAN Control Center",
        "actions": [
            "Configure Approval Request Workflow"
        ]
    },
    {
        "title": "Fluid Fields",
        "actions": [
            "Import Fluid Field Values",
            "Manage Fluid Field Definitions"
        ]
    },
    {
        "title": "Global Core HR",
        "actions": [
            "View Company Org Chart",
            "View Cost Centers",
            "View Enterprise Units",
            "View Legal Entities",
            "View My Org",
            "View Office Locations",
            "View Positions",
            "Create Position",
            "View Job Templates",
            "Add My Mailing Address",
            "Add My Personal Contacts",
            "Add My Physical Address",
            "Configure Associate Classification",
            "Configure Work Authorization Expiration Reminders",
            "Edit Birth Details",
            "Edit My Nationalities",
            "Edit My Personal Demographics",
            "Edit Profile Image",
            "Export Associate Details",
            "Termination ‐ Configuration Settings",
            "Update My Email",
            "Update My Phone Numbers",
            "View My Associate Profile",
            "View My Identifications",
            "View Scope Management",
            "Work Authorization Type Configuration"
        ]
    },
    {
        "title": "Payroll GL Adapter",
        "actions": [
            "Manage General Ledger"
        ]
    },
    {
        "title": "Payroll MyADP Integrations",
        "actions": [
            "View My Direct Deposits",
            "View My Tax Withholding"
        ]
    },
    {
        "title": "Recognition",
        "actions": [
            "Configure Recognitions",
            "Recognize Someone",
            "View My Recognitions"
        ]
    },
    {
        "title": "Recruiting",
        "actions": [
            "View Internal Career Site",
            "Manage Recruiting",
            "View Classic Recruiting Management",
            "Configure Recruiting"
        ]
    },
    {
        "title": "Special Accommodations",
        "actions": [
            "Manage Special Accommodations"
        ]
    },
    {
        "title": "StandOut Assessment",
        "actions": [
            "View My Snapshot"
        ]
    },
    {
        "title": "ATS Connector",
        "actions": [
            "Import New Hires from ATS",
            "Manage New Hires from ATS"
        ]
    },
    {
        "title": "AUTHORIZATION CONTROL CENTER",
        "actions": [
            "Manage Authorizations"
        ]
    },
    {
        "title": "Award Management",
        "actions": [
            "Award Management"
        ]
    },
    {
        "title": "Benefits Supplemental Eligibility and Compliance BMA",
        "actions": [
            "Manage Benefits Supplemental Fields"
        ]
    },
    {
        "title": "Business Processes",
        "actions": [
            "Bulk Job Change",
            "Bulk Termination",
            "Process Internal Hires"
        ]
    },
    {
        "title": "Canada Payroll Front End",
        "actions": [
            "Statement Of Remittance"
        ]
    },
    {
        "title": "Canada Record Of Employment",
        "actions": [
            "Records of Employment"
        ]
    },
    {
        "title": "Career Development",
        "actions": [
            "Configure Career Development",
            "Manage My Career Development"
        ]
    },
    {
        "title": "Career Profile",
        "actions": [
            "Configure Career Profile",
            "Import Education History"
        ]
    },
    {
        "title": "Celergo Integration",
        "actions": [
            "Global Integration Dashboard",
            "Manage global payroll"
        ]
    },
    {
        "title": "Company Links & FAQs",
        "actions": [
            "Company Links and FAQs"
        ]
    },
    {
        "title": "Company Policies",
        "actions": [
            "Manage Company Policies"
        ]
    },
    {
        "title": "Compensation Earnings Import",
        "actions": [
            "Import Pay Data"
        ]
    },
    {
        "title": "Compensation Reviews",
        "actions": [
            "Configure Compensation Reviews",
            "Manage Compensation Reviews"
        ]
    },
    {
        "title": "Competency Management",
        "actions": [
            "Configure Competencies"
        ]
    },
    {
        "title": "Compliance on Demand",
        "actions": [
            "Compliance on Demand"
        ]
    },
    {
        "title": "Core Benefits",
        "actions": [
            "Benefits Data Management",
            "Benefits Implementation",
            "View and Manage Associates’ Benefits",
            "View and Manage Benefits",
            "View Your Benefits",
            "Manage Benefits forms and plan documents"
        ]
    },
    {
        "title": "Core Compensation",
        "actions": [
            "Manage Your Company’s Compensation",
            "View Your Compensation"
        ]
    },
    {
        "title": "DataCloud Configuration",
        "actions": [
            "Reporting Help",
            "Reports & Analytics"
        ]
    },
    {
        "title": "Disciplinary Incidents and Actions",
        "actions": [
            "Disciplinary Incidents",
            "Manage Disciplinary Actions"
        ]
    },
    {
        "title": "Dynamic Teams",
        "actions": [
            "View Dynamic Teams",
            "View Dynamic Teams Chart",
            "View Your Dynamic Teams"
        ]
    },
    {
        "title": "Global Payroll Front End",
        "actions": [
            "View Your Pay",
            "Manage Associates’ Pay Data",
            "Manage Pay Group Transfer",
            "Manage Pay Runs",
            "Manage Payroll Calendar",
            "Manage Payroll Company Data",
            "Manage Settings",
            "Quarter and Year End Checklist",
            "View Payroll Dashboard"
        ]
    },
    {
        "title": "Global Time - NG Leave Adapter",
        "actions": [
            "Manage Company Time Off"
        ]
    },
    {
        "title": "Global Time - WFMgr Adapter",
        "actions": [
            "Manage Time"
        ]
    },
    {
        "title": "Global Time TAM Adapter",
        "actions": [
            "Manage Leave"
        ]
    },
    {
        "title": "Global View Adapter",
        "actions": [
            "Data Reconciliation"
        ]
    },
    {
        "title": "Historical Data Management",
        "actions": [
            "Import Legacy System Data"
        ]
    },
    {
        "title": "HR Dashboard",
        "actions": [
            "View HR Dashboard"
        ]
    },
    {
        "title": "Integration Core",
        "actions": [
            "Integration Core Tooling"
        ]
    },
    {
        "title": "Job Families and Levels",
        "actions": [
            "View Job Families",
            "View Job Chart"
        ]
    },
    {
        "title": "Learning",
        "actions": [
            "Classroom Training Management",
            "Learning Management",
            "My Learning",
            "My Team Learning"
        ]
    },
    {
        "title": "Leave of Absence Core",
        "actions": [
            "Bulk Leave of Absence",
            "Configure Leave of Absence"
        ]
    },
    {
        "title": "Licenses & Certifications",
        "actions": [
            "Manage Licenses & Certifications",
            "View My Licenses & Certifications"
        ]
    },
    {
        "title": "LifeMart",
        "actions": [
            "Discounts"
        ]
    },
    {
        "title": "Lightweight Integrations BMA",
        "actions": [
            "Integrations Dashboard",
            "Lightweight Integrations"
        ]
    },
    {
        "title": "New Hire",
        "actions": [
            "Hiring Dashboard"
        ]
    },
    {
        "title": "News",
        "actions": [
            "Manage News Posts",
            "View Company News"
        ]
    },
    {
        "title": "Payload Handler",
        "actions": [
            "Payload Handler"
        ]
    },
    {
        "title": "Payroll - Autopay Adapter",
        "actions": [
            "Manage prior quarters and year adjustments"
        ]
    },
    {
        "title": "Succession Planning",
        "actions": [
            "Manage Succession Plans",
            "Manage Talent Cycles",
            "Manage/Update Talent Assessments"
        ]
    },
    {
        "title": "System Core - HR",
        "actions": [
            "Associate Directory",
            "Configure Follow‐up Actions",
            "Configure Movement Reasons",
            "Configure Work Relationship Types",
            "Configure Worker Compensation Codes",
            "Organization Management",
            "View People Movements"
        ]
    },
    {
        "title": "System Core Time",
        "actions": [
            "Request Time Off",
            "View Additional Time Features",
            "View and Approve Team Time Off",
            "View and Approve Team Timecards",
            "View and Request Leave of Absence",
            "View My Schedule",
            "View My Time Off",
            "View My Timecard",
            "View Team Schedules",
            "View Time Dashboard"
        ]
    },
    {
        "title": "Talent Management",
        "actions": [
            "Configure Goals / Objectives",
            "Configure Performance Plans",
            "Performance Evaluation",
            "Performance Settings",
            "Performance Setup",
            "Update Goals / Objectives"
        ]
    },
    {
        "title": "Time Off",
        "actions": [
            "Manage Company Time Off"
        ]
    },
    {
        "title": "Timeline Audit",
        "actions": [
            "Audit Reporting"
        ]
    },
    {
        "title": "Union Core HR",
        "actions": [
            "Configure Unions and Bargaining Units"
        ]
    },
    {
        "title": "US Compliance ADP ACA HR Adapter",
        "actions": [
            "Manage ACA Reports"
        ]
    },
    {
        "title": "US Compliance ADP EI9 Adapter",
        "actions": [
            "Manage eI‐9 in ADP eI‐9",
            "Manage eI‐9 Transactions"
        ]
    },
    {
        "title": "US Compliance EEO‐1 Reporting",
        "actions": [
            "Manage EEO Reports"
        ]
    },
    {
        "title": "US Compliance OSHA Reporting",
        "actions": [
            "Manage Illness and Injury Reporting"
        ]
    },
    {
        "title": "US Compliance VETS Reporting",
        "actions": [
            "Manage VETS 4212 Reports"
        ]
    },
    {
        "title": "User Management",
        "actions": [
            "Change Login Methods",
            "Change Product Access",
            "Manage Users",
            "Send Registration Emails",
            "User Management Troubleshooting"
        ]
    },
    {
        "title": "User Registration and Provisioning",
        "actions": [
            "AOID Tracker",
            "Registration Dashboard"
        ]
    },
    {
        "title": "VOE - Voice of the Employee",
        "actions": [
            "Broadcast Message(s) - Voice of the Employee",
            "Manage Your Company Broadcast Message(s)",
            "Manage Your Company Survey(s)",
            "Manage Your Team Broadcast Message(s)",
            "Manage Your Team Survey(s)",
            "Survey(s) - Voice of the Employee"
        ]
    },
    {
        "title": "Wisely Paycard",
        "actions": [
            "View Wisely"
        ]
    }
]

# Expanded helper function to create diverse query variations.
def get_query_variations(title):
    variations = [
        title,
        "I want to know about " + title,
        "How do I use " + title + "?",
        "Tell me more about " + title,
        "What is " + title + "?",
        "Explain " + title,
        "I need information on " + title,
        "I would like details about " + title,
        "Can you help me understand " + title + "?",
        "Give me insights on " + title,
    ]
    # Add specialized query variations for topics related to time off, vacation, or leave.
    lower_title = title.lower()
    if "time off" in lower_title or "vacation" in lower_title or "leave" in lower_title:
        variations += [
            "I want to take a vacation",
            "How do I request time off?",
            "I need to take leave",
            "Show me how to request vacation",
            "Can I get time off?",
            "I want to plan a holiday"
        ]
    return variations

def generate_triplets(catalog_data):
    """
    For each category, generate multiple triplets using different query variations.
    For each action (positive) in the category, sample a negative from a different category.
    """
    triplets = []
    for i, entry in enumerate(catalog_data):
        title = entry["title"]
        variations = get_query_variations(title)
        for query_variant in variations:
            for action in entry["actions"]:
                # Sample a negative from any other category that has actions.
                neg_candidates = [entry2 for j, entry2 in enumerate(catalog_data) if j != i and entry2["actions"]]
                if neg_candidates:
                    neg_entry = random.choice(neg_candidates)
                    negative = random.choice(neg_entry["actions"])
                    triplets.append((query_variant, action, negative))
    return triplets

triplets = generate_triplets(catalog_data)
print("Sample expanded triplets:")
for i in range(min(5, len(triplets))):
    print(triplets[i])
```

```python
def create_vectorizer(all_texts, max_tokens=1000, sequence_length=10):
    """
    Create and adapt a TextVectorization layer on the provided texts.
    """
    vectorizer = TextVectorization(
        max_tokens=max_tokens,
        output_mode="int",
        output_sequence_length=sequence_length,
    )
    vectorizer.adapt(all_texts)
    return vectorizer

# Combine all texts (tile names and actions) for vectorization
catalog = {item['title']:item['actions'] for item in catalog_data}
all_texts = list(catalog.keys()) + [action for actions in catalog.values() for action in actions]
max_tokens = 1000
sequence_length = 10  # Adjust based on your expected text length
vectorizer = create_vectorizer(all_texts, max_tokens, sequence_length)
```

```python
def build_tower(name="tower", sequence_length=10, vocab_size=1000, embedding_dim=16):
    """
    Build a simple tower model:
      - Input layer for tokenized text
      - Embedding layer
      - Global Average Pooling layer
      - Dense layer to produce the final embedding
    """
    input_layer = Input(shape=(sequence_length,), dtype="int32", name=f"{name}_input")
    x = Embedding(input_dim=vocab_size, output_dim=embedding_dim, name=f"{name}_embedding")(input_layer)
    x = GlobalAveragePooling1D(name=f"{name}_pooling")(x)
    x = Dense(embedding_dim, activation="linear", name=f"{name}_dense")(x)
    return Model(input_layer, x, name=name)

# Set embedding dimension and build towers
embedding_dim = 16
query_tower = build_tower("query_tower", sequence_length, max_tokens, embedding_dim)
candidate_tower = build_tower("candidate_tower", sequence_length, max_tokens, embedding_dim)
```

```python
# First, split the triplets into separate lists.
queries, positives, negatives = zip(*triplets)
dataset = tf.data.Dataset.from_tensor_slices((list(queries), list(positives), list(negatives)))

def encode_triplet_with_dummy(q, pos, neg):
    """
    Given a query, positive, and negative string (tf.string scalars), vectorize each string.
    The vectorizer expects a batch of strings so we add a batch dimension and then squeeze it.
    Returns:
      q_enc, pos_enc, neg_enc: tensors of shape [sequence_length] (tf.int64)
      dummy_target: a constant tensor of shape [2] (tf.float32) to serve as a dummy target.
    """
    q_enc = tf.squeeze(vectorizer(tf.expand_dims(q, 0)), axis=0)
    pos_enc = tf.squeeze(vectorizer(tf.expand_dims(pos, 0)), axis=0)
    neg_enc = tf.squeeze(vectorizer(tf.expand_dims(neg, 0)), axis=0)
    dummy_target = tf.constant([0.0, 0.0], dtype=tf.float32)
    return q_enc, pos_enc, neg_enc, dummy_target

def py_encode_triplet(q, pos, neg):
    """Wrapper function to call encode_triplet_with_dummy."""
    return encode_triplet_with_dummy(q, pos, neg)

# Use tf.py_function so that the vectorizer calls execute in eager mode.
dataset = dataset.map(lambda q, pos, neg: tf.py_function(
    func=py_encode_triplet,
    inp=[q, pos, neg],
    Tout=(tf.int64, tf.int64, tf.int64, tf.float32)
))

def set_shapes(q, pos, neg, dummy):
    # Set the expected static shapes.
    q.set_shape([sequence_length])
    pos.set_shape([sequence_length])
    neg.set_shape([sequence_length])
    dummy.set_shape([2])
    return (q, pos, neg), dummy

# Set shapes and batch the dataset.
dataset = dataset.map(set_shapes)
dataset = dataset.batch(32)
```

```python
def build_training_model(query_tower, candidate_tower, sequence_length=10):
    """
    Create a model that:
      - Accepts three inputs (query, positive, negative)
      - Passes them through the appropriate towers
      - Computes dot product similarities between the query and each candidate
      - Concatenates the similarity scores into a two-dimensional output per sample
    """
    # Define inputs for query, positive, and negative texts (already vectorized)
    query_input = Input(shape=(sequence_length,), dtype="int32", name="query_input")
    positive_input = Input(shape=(sequence_length,), dtype="int32", name="positive_input")
    negative_input = Input(shape=(sequence_length,), dtype="int32", name="negative_input")
    
    # Compute embeddings using the towers
    query_emb = query_tower(query_input)
    positive_emb = candidate_tower(positive_input)
    negative_emb = candidate_tower(negative_input)
    
    # Use Lambda layers to compute dot product similarities.
    pos_sim = Lambda(lambda x: tf.reduce_sum(x[0] * x[1], axis=1, keepdims=True))([query_emb, positive_emb])
    neg_sim = Lambda(lambda x: tf.reduce_sum(x[0] * x[1], axis=1, keepdims=True))([query_emb, negative_emb])
    
    # Concatenate the positive and negative similarities.
    output = Lambda(lambda x: tf.concat(x, axis=1))([pos_sim, neg_sim])
    
    return Model(inputs=[query_input, positive_input, negative_input], outputs=output, name="two_tower_training_model")

def triplet_loss(_, y_pred, margin=0.2):
    """
    Custom triplet loss:
    Encourages the positive similarity to be higher than the negative similarity by at least 'margin'.
    """
    pos_sim = y_pred[:, 0]
    neg_sim = y_pred[:, 1]
    loss = tf.maximum(0.0, margin - (pos_sim - neg_sim))
    return tf.reduce_mean(loss)

# Build and compile the training model.
training_model = build_training_model(query_tower, candidate_tower, sequence_length)
training_model.compile(optimizer="adam", loss=triplet_loss)
training_model.summary()
```

```python
# Train the two-tower model for 5 epochs.
training_model.fit(dataset, epochs=5)
```
