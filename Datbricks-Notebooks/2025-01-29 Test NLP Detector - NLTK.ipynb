Filter, Clean and Normalize Data

```python
import nltk
from nltk.corpus import stopwords
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize
from nltk.chunk import ne_chunk

# Download required NLTK data (run once)
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
import re
import string

def normalize_text(text):
    # Replace line breaks (both types) with spaces
    text = re.sub(r'[\n\r]', ' ', text)

    # Replace special characters
    text = re.sub(r'#([[a-zA-ZäöüÄÖÜß]|[0-9]]+)', ' ', text)

    # Replace URLs
    text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ', text)

    # Replace user mentions
    text = re.sub(r'@(\w+)', ' ', text)

    # Convert to lowercase
    text = text.lower()

    # Remove special characters
    # text = re.sub(r'[^a-zA-ZäöüÄÖÜß]', ' ', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Replace two or more subsequent white spaces with a single space
    text = re.sub(r'\ {2,}', ' ', text)

    # Trim white spaces at the beginning or end
    text = text.strip()

    return ' ' if text is None else text
```

```python
# Basic test cases
assert normalize_text("Hello\nWorld") == "hello world"
assert normalize_text("Hello    World") == "hello world"
assert normalize_text("Hello World!") == "hello world"
assert normalize_text("HeLLo WoRLD") == "hello world"
assert normalize_text("     ") == ""
```

```python
import string
from nltk.stem import PorterStemmer, WordNetLemmatizer

def preprocess_text(text):
    # Preprocess the text
    text = normalize_text(text)

    # Tokenize the text
    tokens = text.split()

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Initialize the stemmer and lemmatizer
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    # Stem and lemmatize the filtered tokens
    processed_tokens = [lemmatizer.lemmatize(stemmer.stem(word)) for word in filtered_tokens]
    
    return text, filtered_tokens, processed_tokens
```

```python
# Example usage
# Example usage with assert test cases
text = "Natural Language Processing enables computers to understand human language."

# Expected outputs
expected_original_text = "natural language processing enables computers to understand human language"
expected_filtered_tokens = ['natural', 'language', 'processing', 'enables', 'computers', 'understand', 'human', 'language']
expected_processed_tokens = ['natur', 'languag', 'process', 'enabl', 'comput', 'understand', 'human', 'languag']

# Get the actual outputs from the function
original_text, filtered_tokens, processed_tokens = preprocess_text(text)

# Assert test cases
assert original_text == expected_original_text, f"Expected: {expected_original_text}, but got: {original_text}"
assert filtered_tokens == expected_filtered_tokens, f"Expected: {expected_filtered_tokens}, but got: {filtered_tokens}"
assert processed_tokens == expected_processed_tokens, f"Expected: {expected_processed_tokens}, but got: {processed_tokens}"
```

```python
# Method for model prediction
def predict_POS(text):
    original_text, filtered_tokens, processed_tokens = preprocess_text(text)

    # Tokenize each sentence in the model_input
    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in processed_tokens]

    # Tag each tokenized sentence with POS tags
    tagged_sentences = [nltk.pos_tag(tokens) for tokens in tokenized_sentences]

    # Format the tagged sentences into a string representation
    formatted_sentences = [' '.join([f"{word}_{tag}" for word, tag in sentence]) for sentence in tagged_sentences]

    # Part-of-speech tagging
    pos_tags = nltk.pos_tag(formatted_sentences)
    
    return pos_tags
```
