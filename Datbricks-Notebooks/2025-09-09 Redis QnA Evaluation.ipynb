```python
import boto3
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader


bucket_name = "ml-models-bucket-appbuild-02"
pdf_key = "evaluation_framework/data.pdf"  # S3 key for the PDF
local_path = "data/file.pdf"
s3_client.download_file(bucket_name, pdf_key, local_path)

loader = PyPDFLoader(local_path, headers = None)
documents = loader.load()

# set up the file loader/extractor and text splitter to create chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=2500, chunk_overlap=0
)
chunks = loader.load_and_split(text_splitter)

print("Done preprocessing. Created", len(chunks), "chunks of the original pdf")
```

```python
from redisvl.schema import IndexSchema
from redisvl.index import SearchIndex

index_name = "question_answer_kc"

schema = IndexSchema.from_dict({
  "index": {
    "name": index_name,
    "prefix": index_name,
    "storage_type": "hash"
  },
  "fields": [
        {
        "name": "chunk_id",
        "type": "tag",
        "attrs": {
            "sortable": True
        }
    },
    {"type" : "text", "name" : "reference"},
    {
        "type" : "vector",
        "name" : "embedding",
        "attrs" : {
            "dims": 3072,
            "distance_metric": "cosine",
            "algorithm": "hnsw",
            "datatype": "float32"
        }
    }
  ],
})

index = SearchIndex(schema, redis_client)
index.create(overwrite=True, drop=True)
```

```python
index.load(pdf.to_dict(orient="records"))
```

```python
index.info()['num_docs']
```

```python
from redisvl.query import VectorQuery
from redisvl.query.filter import Tag

user_query = "How do I report a workplace safety concern? | If you observe or become aware of any unsafe condition, you must promptly notify your supervisor and the Human Resources Department so that the Company can take action to correct it."
vector = get_openai_embedding("k9-8Q~xFbsBnXtyDzj0lDnRl9BomhghHjaZQtdfK", user_query)
```

```python
v = VectorQuery(
    vector=np.array(vector).astype(np.float32).tobytes(),
    vector_field_name="embedding",
    num_results=3,
    return_fields=["chunk_id", "reference"],
    return_score=True
)

results = index.query(v)
display(results)
```

```python
# paginate through results
for result in index.paginate(v, page_size=1):
    print(result[0]["chunk_id"], result[0]["vector_distance"], flush=True)
```

```python
# Sort by chunk_id field after vector search limits to topK
v = VectorQuery(
    vector=np.array(vector).astype(np.float32).tobytes(),
    vector_field_name="embedding",
    num_results=3,
    return_fields=["chunk_id", "reference"],
    return_score=True
)

# Decompose vector_query into the core query and the params
query = v.query
params = v.params

# Pass query and params direct to index.search()
result = index.search(
    query.sort_by("chunk_id", asc=True),
    params
)

pd.DataFrame([doc.__dict__ for doc in result.docs])
```
